```python
import ROOT
from ROOT import gStyle
gStyle.SetOptStat(0)
```

```python
myFile_gel = ROOT.TFile.Open("/Users/bertabeltran/muon_data/data/new_desktop/DoubleCoincidence_8ns_OpticalGel/FILTERED/HcompassF_DoubleCoincidence_8ns_OpticalGel_20231005_095508.root")
```

```python
myFile_gel.ls()
```

```python
dir_gel = myFile_gel.GetDirectory("Energy");
```

```python
dir_gel.ls()
```

```python
hist_gel_0 = dir_gel.Get("_F_EnergyCH0@DT5751_1615;1")
hist_gel_0.Rebin(7)
c0 = ROOT.TCanvas("c0"," ",800,600)

hist_gel_0.Fit("gaus", "", "",1000,1700)
gfit_gel_0= hist_gel_0.GetFunction("gaus")
mgel0=gfit_gel_0.GetParameter(1)
mgel0_err= gfit_gel_0.GetParError(1)

hist_gel_0.Draw()
c0.Draw()
```

```python
hist_gel_1 = dir_gel.Get("_F_EnergyCH1@DT5751_1615;1")
hist_gel_1.Rebin(7)

c0.Clear()
hist_gel_1.Fit("gaus", "", "",900,1500)
gfit_gel_1= hist_gel_1.GetFunction("gaus")
mgel1=gfit_gel_1.GetParameter(1)
mgel1_err= gfit_gel_1.GetParError(1)

hist_gel_1.Draw()
c0.Draw()
```

```python
myFile_pad = ROOT.TFile.Open("/Users/bertabeltran/muon_data/data/new_desktop/DoubleCoincidence_8ns_SiliconPad/FILTERED/HcompassF_DoubleCoincidence_8ns_SiliconPad_20231006_100744.root")
dir_pad = myFile_pad.GetDirectory("Energy");
dir_pad.ls()
hist_pad_0 = dir_pad.Get("_F_EnergyCH0@DT5751_1615")
hist_pad_0.Rebin(7)

c0.Clear()
hist_pad_0.Fit("gaus", "", "",1000,1600)
gfit_pad_0= hist_pad_0.GetFunction("gaus")
mpad0=gfit_pad_0.GetParameter(1)
mpad0_err= gfit_pad_0.GetParError(1)




hist_pad_0.Draw()
c0.Draw()
```

```python
hist_pad_1 = dir_pad.Get("_F_EnergyCH1@DT5751_1615")
hist_pad_1.Rebin(7)

c0.Clear()
hist_pad_1.Fit("gaus", "", "",900,1500)
gfit_pad_1= hist_pad_1.GetFunction("gaus")
mpad1=gfit_pad_1.GetParameter(1)
mpad1_err= gfit_pad_1.GetParError(1)


hist_pad_1.Draw()
c0.Draw()

```

```python

c = ROOT.TCanvas("myCanvasName","Scintillator with waveshifting fibre",800,1000)

gStyle.SetOptStat(0)
# channel 0 pad 

pad1 = ROOT.TPad("pad1","pad1",0,0.5,1,1) 
pad1.Draw()
pad1.cd()

hist_gel_0.SetFillColor(0)
#hist_bkg.GetYaxis().SetRangeUser(0, 0.05)#
hist_gel_0.SetNormFactor(1)
hist_gel_0.Draw()
hist_gel_0.GetYaxis().SetTitleOffset(1.5) #1.2


hist_pad_0.SetLineColor(1)
hist_pad_0.SetFillColor(0)
hist_pad_0.SetNormFactor(1)
hist_pad_0.Draw("same")
#c.SetLogy()


leg = ROOT.TLegend(.7,.8,.9,.9)
leg.AddEntry(hist_gel_0,"Optical gel ","L")
leg.AddEntry(hist_pad_0,"Silicon pad ","L")
leg.Draw()

tex=ROOT.TLatex(0.25,0.93," Channel 0 ");
tex.SetNDC();
tex.SetTextSize(0.04);
tex.Draw();

gel_mean_0 = 'Gel mean {:.1f} #pm {:.2f}'.format(mgel0,mgel0_err)
print(gel_mean_0)

tex_gel_fit0=ROOT.TLatex(0.5,0.7,gel_mean_0);
tex_gel_fit0.SetNDC();
tex_gel_fit0.SetTextSize(0.04);
tex_gel_fit0.Draw();


pad_mean_0 = 'Pad mean {:.1f} #pm {:.2f}'.format(mpad0,mpad0_err)
print(pad_mean_0)


tex_pad_fit0=ROOT.TLatex(0.5,0.63,pad_mean_0);
tex_pad_fit0.SetNDC();
tex_pad_fit0.SetTextSize(0.04);
tex_pad_fit0.Draw();

change0='Change {:.2f}%'.format(((mgel0-mpad0)*100)/mgel0)

tex_change0=ROOT.TLatex(0.55,0.56,change0);
tex_change0.SetNDC();
tex_change0.SetTextSize(0.04);
tex_change0.Draw();

# channel 1 pad 


c.cd()
pad2 = ROOT.TPad("pad2","pad2",0,0,1,0.5) 
pad2.Draw()
pad2.cd()

hist_gel_1.SetFillColor(0)
#hist_bkg.GetYaxis().SetRangeUser(0, 0.05)#
hist_gel_1.SetNormFactor(1)
hist_gel_1.Draw()
hist_gel_1.GetYaxis().SetTitleOffset(1.5) #1.2

hist_pad_1.SetLineColor(1)
hist_pad_1.SetFillColor(0)
hist_pad_1.SetNormFactor(1)
hist_pad_1.Draw("same")
#c.SetLogy()
gStyle.SetOptStat(0)

leg1 = ROOT.TLegend(.7,.8,.9,.9)
leg1.AddEntry(hist_gel_0,"Optical gel ","L")
leg1.AddEntry(hist_pad_0,"Silicon pad ","L")
leg1.Draw()

tex1=ROOT.TLatex(0.25,0.93," Channel 1 ");
tex1.SetNDC();
tex1.SetTextSize(0.04);
tex1.Draw();


gel_mean_1 = 'Gel mean {:.1f} #pm {:.2f}'.format(mgel1,mgel1_err)
print(gel_mean_1)

tex_gel_fit1=ROOT.TLatex(0.5,0.7,gel_mean_1);
tex_gel_fit1.SetNDC();
tex_gel_fit1.SetTextSize(0.04);
tex_gel_fit1.Draw();


pad_mean_1 = 'Pad mean {:.1f} #pm {:.2f}'.format(mpad1,mpad1_err)
print(pad_mean_1)


tex_pad_fit1=ROOT.TLatex(0.5,0.63,pad_mean_1);
tex_pad_fit1.SetNDC();
tex_pad_fit1.SetTextSize(0.04);
tex_pad_fit1.Draw();

change1='Change {:.2f}%'.format(((mgel1-mpad1)*100)/mgel1)

tex_change1=ROOT.TLatex(0.55,0.56,change1);
tex_change1.SetNDC();
tex_change1.SetTextSize(0.04);
tex_change1.Draw();


c.Draw()
c.SaveAs("/Users/bertabeltran/muon_data/data/OpticalGel_vs_SiliconPad.png");
```
