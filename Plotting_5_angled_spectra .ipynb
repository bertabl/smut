```python
from array import array
c1 = ROOT.TCanvas("c1"," ",800,600)

n = 6;
x  = array( 'f', [0, 1, 2, 3, 4, 5] )
ex = array( 'f', [0,0,0,0,0,0] )
y  = array( 'f', [m0,m1,m2,m3,m4,m5])
ey = array( 'f', [m0_err,m1_err,m2_err,m3_err,m4_err,m5_err])
 
gr = ROOT.TGraphErrors(n,x,y,ex,ey)
gr.SetTitle("Mean of energy spectrum vs angle;Angle (deg); Mean of energy spectrum (ADC counts)")
gr.SetMarkerColor(4)
gr.SetMarkerStyle(21)
gr.Draw("ALP")
c1.Draw()
c1.SaveAs("/Users/bertabeltran/muon_data/data/SiPM_angles_graph.png");
```
