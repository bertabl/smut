```python
import ROOT
from ROOT import gStyle
gStyle.SetOptStat(0)
```

```python
c=ROOT.TCanvas("c0"," ",800,600)
func = ROOT.TF1("func", "(1/(1+ TMath::Exp(-[0]*(x-[1]))))", 0, 600) 
func.SetParameters(0.04, 180)
func.Draw()
c.Draw()
```

```python
#lower + (upper - lower)/(1 + Exp[a + b x]),
c1=ROOT.TCanvas("c1"," ",800,600)
func1 = ROOT.TF1("func1", "([0] + ([1] - [0])/(1+ TMath::Exp(-[2]*(x-[3]))))", 0, 600) 
func1.SetParameters(0.4,8, 0.04, 180)
func1.Draw()
c1.Draw()
```

```python
#Asimetric sigmoid 

c2=ROOT.TCanvas("c2"," ",800,600)
func2 = ROOT.TF1("func2", "(0.5)*((TMath::Power((1+TMath::Exp((-1/[0])*(x-[1]-[0]*TMath::Log(TMath::Power(2,(1/[2]))-1)))),-[2]))+(1-TMath::Power((1+TMath::Exp((1/[0])*(x-[1]+[0]*TMath::Log(TMath::Power(2,(1/[3]))-1)))),-[3])))", 0, 600) 
func2.SetParameters(25,200,200,0.3)
func2.Draw()
c2.Draw()
```

```python
#toe sectionn 
c3=ROOT.TCanvas("c3"," ",800,600)
func3 = ROOT.TF1("func3", "TMath::Power((1+TMath::Exp((-1/[0])*(x-[1]-[0]*TMath::Log(TMath::Power(2,(1/[2]))-1)))),-[2])", 0, 600) 
#func3.SetParameters(30,120,1.25383e+06)
func3.SetParameters(30,120,1)
func3.Draw()

func4 = ROOT.TF1("func4", "TMath::Power((1+TMath::Exp((-1/[0])*(x-[1]-[0]*TMath::Log(TMath::Power(2,(1/[2]))-1)))),-[2])", 0, 600) 
func4.SetParameters(30,120,1.25383e+06)
#func4.SetParameters(30,120,1)
func4.SetLineColor(3)
func4.Draw("same")

c3.Draw()
```

```python
#shoulder secction 
c4=ROOT.TCanvas("c4"," ",800,600)
func4 = ROOT.TF1("func4", "1-TMath::Power((1+TMath::Exp((1/[0])*(x-[1]+[0]*TMath::Log(TMath::Power(2,(1/[2]))-1)))),-[2])", 0, 600) 
func4.SetParameters(25,200,1)
func4.Draw()
c4.Draw()
```
